```python
dataset = pd.read_csv('train.csv')
dataset2 = pd.read_csv('test.csv')
dataset.drop(['Id'],axis=1,inplace=True)
```

```python
for features in dataset.columns:
    if dataset[features].isnull().sum()>1:
        if dataset[features].dtypes !='O':
            dataset[features].fillna(dataset[features].mean(),inplace=True) 
```

```python
dataset.isnull().sum()
```

```python
categorical_features = []
numerical_features = []
for features in dataset.columns:
    if dataset[features].dtypes !='O':
        numerical_features.append(features)
    else:
        categorical_features.append(features)
```

```python
def replace_cat_feature(dataset,features_nan):
    data=dataset.copy()
    data[features_nan]=data[features_nan].fillna('Missing')
    return data
```

```python
dataset=replace_cat_feature(dataset,categorical_features)
```

```python
dataset = dataset[['MSSubClass', 'MSZoning', 'Neighborhood', 'OverallQual', 'YearRemodAdd',
       'RoofStyle', 'BsmtQual', 'BsmtExposure', 'HeatingQC', 'CentralAir',
       '1stFlrSF', 'GrLivArea', 'BsmtFullBath', 'KitchenQual', 'Fireplaces',
       'FireplaceQu', 'GarageType', 'GarageFinish', 'GarageCars', 'PavedDrive',
       'SaleCondition','SalePrice']]
```
